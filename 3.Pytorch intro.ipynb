### Creating tensors in pytorch


```python
import numpy as np
import torch
```

```python
#1. We can create tensor from a list or a numpy array.
np.set_printoptions(precision=3)
a = [1, 2, 3]
b = np.array([4, 5, 6], dtype=np.int32)
t_a = torch.tensor(a)
t_b = torch.from_numpy(b)
print(t_a)
print(t_b)
```

```python
t_ones = torch.ones([2,3]) # a matrix of 1's
print(t_ones.shape) # shape of a tensor
print(t_ones)
```

```python
rand_tensor = torch.rand(2,3)
print(rand_tensor)
```

### Manipulating data type and shape of a tensor


```python
t_a_new = t_a.to(torch.int64)
print(t_a_new.dtype)
```

Certain operations require tensors to have a certain number of dimension(this is, rank), associated with a certain number of elements(shape). Thus, we need to change the shape of a tensor, add a new dimension.

```python
## Transposing a tensor
t = torch.rand([3,5])
t_tr = torch.transpose(t, 0, 1) #swap dimension 0 and dimension 1.
print(t.shape,' --> ',t_tr.shape)
```

```python
## Reshaping a tensor
t = torch.zeros(30)
t_reshape = t.reshape(5,6)
print(t_reshape.shape)
```

```python
## Removing unnecessary dimensions (dimensions that have size 1)
t = torch.zeros(1, 2, 1, 4, 1)
t_sqz = torch.squeeze(t, 2)
print(t.shape, ' --> ', t_sqz.shape)
```

### Applying mathematical operatorions to tensors

```python
torch.manual_seed(1)
t1 = 2 * torch.rand(5,2) - 1 # generate random numbers from normal distribution N(0,1)
t2 = torch.normal(mean=0, std = 1, size = (5,2))

# to calculate element-wise product of t1 and t2
t3 = torch.multiply(t1, t2)
print(t3)
```

```python
# Calculating some statistics along an axis
t4 = torch.mean(t1, axis = 0)
print(t4)
print(torch.sum(t2, axis = 0))
print(torch.std(t2))
```

```python
## Matrix multiplication
t5 = torch.matmul(t1, torch.transpose(t2, 0, 1))
print(t5)
```

```python
t5 = torch.matmul(torch.transpose(t1, 0, 1), t2)
print(t5)
```

```python
## Calculating Lp norm of a matrix
norm_t1 = torch.linalg.norm(t1, ord = 2, dim = 1)
print(norm_t1)
```
